#Missing Values


```
import numpy as np
import pandas as pd
```

```
#create a dataframe

df = pd.DataFrame({'A':[1,2,np.nan],
                   'B':[6,np.nan,np.nan],
                   'C':[1,2,3]})

df
```

```
df.isna().sum()
```

```
#removing the NaN values

df.dropna()
```

```
df
```

```
df.dropna(axis= 1)
```

```
df.fillna(value = 'Mean or median or mode of data')
```

```
df['A'].fillna(value = df['A'].mean())
```
