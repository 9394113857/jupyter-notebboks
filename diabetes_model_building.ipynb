```python
zero_columns = df.eq(0).any() # any() means "at least one is True" or "any of the elements satisfy the condition."
print(zero_columns) # True means atleast 1 zero value, False means atleast no zero values found
```

**1.	df:** is assumed to be a Pandas DataFrame.

**2.	df.eq(0):** performs element-wise comparison of the DataFrame df with the value 0. It creates a new DataFrame with the same shape as df but with boolean values (True where the original value was 0, and False where it was not).

**3.	.any():** is then applied to the boolean DataFrame, which returns a Series indicating whether any True value exists in each column. If at least one True value exists in a column, it means that the column contains at least one zero.

**4.Finally, the print(zero_columns)** statement will output the Series showing True for columns containing at least one zero, and False for columns with no zeros.

```python
# Replacing NaN value by mean, median depending upon distribution
df_copy['Glucose'].fillna(df_copy['Glucose'].mean(), inplace=True)
df_copy['BloodPressure'].fillna(df_copy['BloodPressure'].mean(), inplace=True)
df_copy['SkinThickness'].fillna(df_copy['SkinThickness'].median(), inplace=True)
df_copy['Insulin'].fillna(df_copy['Insulin'].median(), inplace=True)
df_copy['BMI'].fillna(df_copy['BMI'].median(), inplace=True)
```

```python
# After replaced all the NaN values in the specified columns with specified conditions:
import pandas as pd

# Assuming you have the DataFrame df_copy

# Filter rows with NaN values and save them to a CSV file
rows_with_nan = df_copy[df_copy.isnull().any(axis=1)]
rows_with_nan.to_csv("rows_without_nan.csv", index=False)

# Read the CSV file into a DataFrame
df_from_csv = pd.read_csv("rows_without_nan.csv")

# Get the count of NaN values in each column from the CSV file
nan_counts_from_csv = df_from_csv.isnull().sum()

# Display the count of NaN values along with the column names
print("\033[1mColumn-wise NaN values counts from CSV:\033[0m")
print(nan_counts_from_csv)
```

```python
# Model Building
from sklearn.model_selection import train_test_split
X = df.drop(columns='Outcome')
Y = df['Outcome']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=0)
```

```python
# Creating Random Forest Model
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=20)
classifier.fit(X_train, y_train)
```

```python
# Load the dumped model
filename = 'diabetes-prediction-rfc-model.pkl'
loaded_model = pickle.load(open(filename, 'rb'))

# Take user input
user_input = []
pregnancies = int(input("Number of pregnancies: "))
user_input.append(pregnancies)
glucose = int(input("Glucose level: "))
user_input.append(glucose)
blood_pressure = int(input("Blood pressure: "))
user_input.append(blood_pressure)
skin_thickness = int(input("Skin thickness: "))
user_input.append(skin_thickness)
insulin = int(input("Insulin level: "))
user_input.append(insulin)
bmi = float(input("BMI: "))
user_input.append(bmi)
dpf = float(input("Diabetes pedigree function: "))
user_input.append(dpf)
age = int(input("Age: "))
user_input.append(age)

# Create a DataFrame from user input
input_df = pd.DataFrame([user_input], columns=['Pregnancies', 'Glucose', 'BloodPressure',
                                               'SkinThickness', 'Insulin', 'BMI', 'DPF', 'Age'])

# Make predictions
prediction = loaded_model.predict(input_df)

# Display the prediction
if prediction[0] == 0:
    print("The person is not diabetic.")
else:
    print("The person is diabetic.")
```
