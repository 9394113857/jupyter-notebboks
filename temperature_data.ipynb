```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Task 1: Load the data from the CSV file into a Pandas DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/9394113857/Data-Sets/raghu/temp.csv')
```

```python
df.head()
```

```python
# Task 2: Calculate the average temperature for each city
average_temp = df.groupby('City')['Temperature_Celsius'].mean()
```

```python
# Task 3: Find the maximum temperature recorded and the corresponding date for each city
max_temp_info = df.groupby('City').apply(lambda x: x.loc[x['Temperature_Celsius'].idxmax()])
```

Let's explore some additional plots and the insights we can gain from them:

**1.Line Plot:**

A line plot can be used to visualize the temperature trend over time for a specific city. This plot will show how the temperature changes day by day.

Insights:

Identify seasonal patterns in temperature changes.
Observe any upward or downward trends in temperature over time.
Identify sudden spikes or drops in temperature that might indicate extreme weather events.

**2.Scatter Plot:**

A scatter plot can be used to explore the relationship between two variables, such as temperature and date. Each data point represents a specific temperature reading for a given date.

Insights:

Identify any correlation between date and temperature.
Check if there are any outliers or unusual temperature readings.

**3.Box Plot:**

A box plot can be used to visualize the distribution of temperatures for each city. It shows the median, quartiles, and potential outliers in the data.

Insights:

Compare temperature distributions across different cities.
Identify cities with higher temperature variations (larger interquartile range).
Spot potential outliers that may represent extreme weather conditions.

**4.Heatmap:**

A heatmap can be used to visualize the temperature variations across multiple cities and dates. The colors represent the temperature values, making it easy to spot patterns.

Insights:

Identify which cities have consistently higher or lower temperatures.
Observe how temperature changes with time for each city.
Spot any correlations between temperatures of different cities.

**5.Histogram:**

A histogram can be used to understand the frequency distribution of temperatures. It shows how many times each temperature range occurs.

Insights:

Observe the most common temperature range for each city.
Check if temperature distributions follow a normal or skewed pattern.
Identify temperature ranges that occur rarely or not at all.

**6.Area Plot:**

An area plot can be used to compare temperature trends for multiple cities over time. Each city's temperature is represented by a colored area on the chart.

Insights:

Compare temperature trends between different cities.
Identify cities with similar or contrasting temperature patterns.

```python
# Task 4: Visualize the average temperature for each city using a bar chart
plt.figure(figsize=(10, 6))
plt.bar(average_temp.index, average_temp.values, color='skyblue')
plt.xlabel('City')
plt.ylabel('Average Temperature (Celsius)')
plt.title('Average Temperature for Each City')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Line Plot (Temperature trend over time for a specific city - New York):

```python
# Assuming you have already loaded the data into the 'df' DataFrame

# Filter data for New York city
new_york_data = df[df['City'] == 'New York']

# Plot the temperature trend
plt.figure(figsize=(10, 6))
plt.plot(new_york_data['Date'], new_york_data['Temperature_Celsius'], marker='o', color='b')
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature Trend in New York')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Scatter Plot (Relationship between temperature and date):

```python
# Assuming you have already loaded the data into the 'df' DataFrame

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Date'], df['Temperature_Celsius'], marker='o', color='r')
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature Scatter Plot')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Box Plot (Distribution of temperatures for each city):

```python
# Assuming you have already loaded the data into the 'df' DataFrame

# Plot the box plot
plt.figure(figsize=(10, 6))
plt.boxplot([df[df['City'] == 'New York']['Temperature_Celsius'],
             df[df['City'] == 'San Francisco']['Temperature_Celsius'],
             df[df['City'] == 'Seattle']['Temperature_Celsius']],
            labels=['New York', 'San Francisco', 'Seattle'])
plt.xlabel('City')
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature Distribution for Each City')
plt.tight_layout()
plt.show()
```

Heatmap (Temperature variations across multiple cities and dates):

```python
import seaborn as sns

# Assuming you have already loaded the data into the 'df' DataFrame

# Pivot the DataFrame to create a heatmap
heatmap_data = df.pivot('City', 'Date', 'Temperature_Celsius')

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt=".1f")
plt.xlabel('Date')
plt.ylabel('City')
plt.title('Temperature Variations (Heatmap)')
plt.tight_layout()
plt.show()
```

Histogram (Frequency distribution of temperatures):

```python
# Assuming you have already loaded the data into the 'df' DataFrame

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Temperature_Celsius'], bins=10, color='purple', alpha=0.7)
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Frequency')
plt.title('Temperature Histogram')
plt.tight_layout()
plt.show()
```

Area Plot (Compare temperature trends for multiple cities over time):

```python
# Assuming you have already loaded the data into the 'df' DataFrame

# Pivot the DataFrame to create an area plot
area_plot_data = df.pivot('Date', 'City', 'Temperature_Celsius')

# Plot the area plot
plt.figure(figsize=(10, 6))
plt.stackplot(area_plot_data.index, area_plot_data.values.T, labels=area_plot_data.columns)
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature Trend for Each City')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
